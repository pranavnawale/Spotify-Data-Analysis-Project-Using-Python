```python
# install the dependencies
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
```

```python
sorted_tracks=tracks.sort_values("popularity",ascending=True)
sorted_tracks.head()
```

```python
# Top_10 Song popular in 90
top_10_90=tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=True)
top_10_90.head(10)
```

```python
tracks.set_index("release_date",inplace=True)
tracks.index=pd.to_datetime(tracks.index)
tracks.head()
```

```python
# Artists in our Dataset
tracks["artists"].value_counts()
```

```python
# Song are in Millseconds they are converted into seconds
tracks["duration"]=tracks["duration_ms"].apply(lambda x:round(x/1000))
tracks.drop("duration_ms",axis=1,inplace=True)
```

```python
tracks.duration.head()
```

## Visualization

```python
# Heatmap
corr=tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(16,8))
heatmap=sns.heatmap(corr,annot=True,fmt=".1g",vmin=1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Purple")
heatmap.set_title("Correlatiobn Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_data=tracks.sample(int(0.004*len(tracks)))
sample_data.head()
```

```python
print(len(sample_data))
```

```python
plt.figure(figsize=(10,8))
sns.regplot(data=sample_data,x='energy',y='loudness',color='c')
plt.title("Loundness Vs Energy Coreelation")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.regplot(data=sample_data,x='acousticness',y='popularity',color='red')
plt.title("acousticness Vs popularity Coreelation")
plt.show()
```

```python
# Add new column dates
tracks["Dates"]=tracks.index.get_level_values('release_date')
tracks.Dates=pd.to_datetime(tracks.Dates)
years=tracks.Dates.dt.year
```

```python
# Distriburion Plot
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist")
plt.title("Number Of Song per year")
plt.show()
```

```python
total_duration=tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_duration,ax=ax,errwidth=False)
plt.title("Year vs Duration")
plt.xticks(rotation=90)
```
